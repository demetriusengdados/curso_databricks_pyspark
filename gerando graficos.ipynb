```
%sql
select pais, sum(preco) as total_vendido from vinho1
where preco > 0
GROUP by pais
ORDER by total_vendido desc
limit 10
```

```
%sql 
select pais, variante, sum(preco) as total_vendido from vinho1
where preco > 0
group by pais, variante
order by total_vendido desc
limit 10
```
